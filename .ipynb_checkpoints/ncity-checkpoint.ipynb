```python
def albumTracks(artist):
    tracks = []
    for album in sp.artist_albums(artist)['items']:
        album_uri = album['uri']
        
        for track in sp.album_tracks(album_uri)['items']:
            tracks.append(track['uri'])
            
    base_df = pd.DataFrame()
    num = 0
    for x in range(50, len(tracks), 50):
        num = x
        df = pd.DataFrame(sp.audio_features(tracks[(num - 50):num]))
        base_df = pd.concat([base_df, df])
        
    final_df = pd.DataFrame(sp.audio_features(tracks[num:len(tracks) + 1]))
    final_df = pd.concat([base_df, final_df])
            
    return final_df
```

```python
nct_dfs = []
for group in [nct127, nctdream, wayv, nctu, nct]:
    nct_dfs.append(albumTracks(group))
```

```python
def add_columns(df):
    df['artist_name'] = df['uri'].apply(lambda x: sp.track(x)['artists'][0]['name'])
    df['album'] = df['uri'].apply(lambda x: sp.track(x)['album']['name'])
    df['track_name'] = df['uri'].apply(lambda x: sp.track(x)['name'])
    df['popularity'] = df['uri'].apply(lambda x: sp.track(x)['popularity'])
    
    return df
```

```python
for df_num in range(len(nct_dfs)):
    nct_dfs[df_num] = add_columns(nct_dfs[df_num])
```

```python
nct_df = pd.concat(nct_dfs).reset_index().drop(columns='index')
```

```python
nct_df['if_nct'] = nct_df['artist_name'].apply(lambda x: True if 'NCT' in x else (True if 'WayV' in x else False))
nct_df = nct_df[nct_df['if_nct'] == True].reset_index().drop(columns=['index', 'if_nct'])
```

```python
nct_df = nct_df.drop_duplicates(subset='track_name').reset_index().drop(columns='index')
nct_df['no_alt'] = nct_df['track_name'].apply(lambda x: True if 'live' in str.lower(x) else (True if 'remix' in str.lower(x) else (True if 'instrumental' in str.lower(x) else False)))
nct_df = nct_df[nct_df['no_alt'] == False].reset_index().drop(columns=['index', 'no_alt'])
```

```python
# change NCT songs to NCT U 
nct_df['artist_name'] = nct_df['artist_name'].apply(lambda x: 'NCT U' if x == 'NCT' else x)
nct_df = nct_df.drop(labels=[178,179,180]) # removes without you as already listed in empathy album
nct_df = nct_df.reset_index(drop=True)
```

```python
def filter_by_NCT(df):
    nct_filter = df['Artist'].apply(lambda x: True if 'NCT' in x else (True if 'WayV' in x else False))
    df['Is NCT'] = nct_filter
    
    return df[df['Is NCT'] == True]

def numAlbumCharts(df, year):
    artist_df = df[df['Year Released'] == year].reset_index(drop=True)
    
    return filter_by_NCT(artist_df)
```

```python
import requests 
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://en.wikipedia.org/wiki/NCT_discography")
soup = BeautifulSoup(page.content, 'html.parser')
wiki_album_data = soup.find_all(class_="wikitable plainrowheaders")
```

```python
def wiki_album_titles(album):
    try:
        return album.find("a")['title']
    except: 
        return album.find("i").get_text()

def wiki_chart_data(data, unit, isAlbum):
    if isinstance(data, list) == True :
        album_titles = [wiki_album_titles(album) for album in data]
        korPeaks = [album.find_all("td")[1].get_text() for album in data]
        canPeaks = [album.find_all("td")[2].get_text() for album in data]
        jpnPeaks = [album.find_all("td")[4].get_text() for album in data]
        if isAlbum == True:
            ukPeaks = [album.find_all("td")[7].get_text() for album in data]
            usPeaks = [album.find_all("td")[8].get_text() for album in data]
            usWorldPeaks = [album.find_all("td")[10].get_text().strip() for album in data]
        else: 
            ukPeaks = [album.find_all("td")[6].get_text() for album in data]
            usPeaks = [album.find_all("td")[7].get_text() for album in data]
            usWorldPeaks = [album.find_all("td")[9].get_text().strip() for album in data]
    else:
        album_titles = [wiki_album_titles(data)]
        korPeaks = [data.find_all("td")[1].get_text()]
        canPeaks = [data.find_all("td")[2].get_text()]
        jpnPeaks = [data.find_all("td")[4].get_text()]
        if isAlbum == True:
            ukPeaks = [data.find_all("td")[7].get_text()]
            usPeaks = [data.find_all("td")[8].get_text()]
            usWorldPeaks = [data.find_all("td")[10].get_text().strip()]
        else: 
            ukPeaks = [data.find_all("td")[6].get_text()]
            usPeaks = [data.find_all("td")[7].get_text()]
            usWorldPeaks = [data.find_all("td")[9].get_text().strip()]
    
    df = pd.DataFrame({
        "Album/EP": album_titles,
        "Unit": unit,
        "KOR Peak": korPeaks,
        "CAN Peak": canPeaks,
        "JPN Peak": jpnPeaks,
        "UK Peak": ukPeaks,
        "US Peak": usPeaks,
        "US World Peak": usWorldPeaks
    })
    
    return df

def combine_wikidfs(charts):
    wikiNCT_albums = [wiki_album_data[0].find_all("tr")[2:5], # nct u
                      wiki_album_data[0].find_all("tr")[6:9], # nct 127
                      wiki_album_data[0].find_all("tr")[11], # nct dream
                      wiki_album_data[0].find_all("tr")[13]] # wayv
    wikiNCT_eps = [wiki_album_data[4].find_all("tr")[3:10], # nct 127
                   wiki_album_data[4].find_all("tr")[11:16], # nct dream
                   wiki_album_data[4].find_all("tr")[17:20]] # nct wayv
    
    df_list = []
    unit_list = ["NCT U", "NCT 127", "NCT Dream", "WayV"]
    
    for x in range(len(wikiNCT_albums)):
        df_list.append(wiki_chart_data(wikiNCT_albums[x], unit_list[x], True))
        
    for x in range(len(wikiNCT_eps)):
        df_list.append(wiki_chart_data(wikiNCT_eps[x], unit_list[x + 1], False))
        
    final_df = pd.concat(df_list).reset_index(drop=True)
    
    # fix specific cells
    final_df.loc[1,"Album/EP"] = "NCT 2020 Resonance Pt. 1"
    final_df.loc[19,"JPN Peak"] = "2"
    final_df.loc[19,"US World Peak"] = "7"
    
    return final_df
```

```python
def get_ytplaylist(playlist, pageToken, videoItems):
    if pageToken == 'finished':
        return []
    else:
        if videoItems == False:
            videoItems = []
            request = youtube.playlistItems().list(
                    part="contentDetails, snippet",
                    playlistId=playlist,
                    maxResults="50"
                )
            response = request.execute()

            for video in response['items']:
                videoId = video['contentDetails']['videoId']
                videoItems.append(videoId)

            return videoItems + get_ytplaylist(playlist, response['nextPageToken'], True)
        else:
            videoItems = []
            request = youtube.playlistItems().list(
                    part="contentDetails, snippet",
                    playlistId=playlist,
                    maxResults="50",
                    pageToken=pageToken
                )
            response = request.execute()

            for video in response['items']:
                videoId = video['contentDetails']['videoId']
                videoItems.append(videoId)

            if ('nextPageToken' in response) == True: 
                return videoItems + get_ytplaylist(playlist, response['nextPageToken'], True)
            elif ('nextPageToken' in response) == False:
                return videoItems + get_ytplaylist(playlist, 'finished', True)
```

```python
def get_ytvideo_info(videoID_df): 
    length = len(videoID_df)
    channelTitles = []
    dates = []
    videoTitles = []
    viewsL = []
    likesL = []
    dislikesL = []
    
    for x in range(0, length, 50):
        request = youtube.videos().list(
            part="contentDetails, statistics, snippet",
            id=list(videoID_df[x:x+50])
        )
        response = request.execute()
        
        for video in response['items']:
            channelTitles.append(video['snippet']['channelTitle'])
            dates.append(video['snippet']['publishedAt'])
            videoTitles.append(video['snippet']['title'])
            viewsL.append(video['statistics']['viewCount'])
            try:
                likesL.append(video['statistics']['likeCount'])
                dislikesL.append(video['statistics']['dislikeCount'])
            except:
                likesL.append(None)
                dislikesL.append(None)
            
    if x != length:
        request = youtube.videos().list(
            part="contentDetails, statistics, snippet",
            id=list(videoID_df[x:])
        )
        response = request.execute()
        
        for video in response['items']:
            channelTitles.append(video['snippet']['channelTitle'])
            dates.append(video['snippet']['publishedAt'])
            videoTitles.append(video['snippet']['title'])
            viewsL.append(video['statistics']['viewCount'])
            try:
                likesL.append(video['statistics']['likeCount'])
                dislikesL.append(video['statistics']['dislikeCount'])
            except:
                likesL.append(None)
                dislikesL.append(None)
        
    df = pd.DataFrame({
        "Channel": channelTitles,
        "Video Title": videoTitles,
        "Date Published": dates,
        "Views": viewsL,
        "Likes": likesL,
        "Dislikes": dislikesL
    })
    
    return df
```
